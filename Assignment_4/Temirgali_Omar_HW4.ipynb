```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
day = pd.read_csv("day.csv")
day.shape
```

Dataset characteristics	
-day.csv - bike sharing counts aggregated on daily basis. Records: 731 days
day.csv have the following fields:
	- instant: record index
	- dteday : date
	- season : season (1:springer, 2:summer, 3:fall, 4:winter)
	- yr : year (0: 2011, 1:2012)
	- mnth : month ( 1 to 12)
	- holiday : weather day is holiday or not (extracted from http://dchr.dc.gov/page/holiday-schedule)
	- weekday : day of the week
	- workingday : if day is neither weekend nor holiday is 1, otherwise is 0.
	+ weathersit : 
		- 1: Clear, Few clouds, Partly cloudy, Partly cloudy
		- 2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist
		- 3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds
		- 4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog
	- temp : Normalized temperature in Celsius. The values are divided to 41 (max)
	- atemp: Normalized feeling temperature in Celsius. The values are divided to 50 (max)
	- hum: Normalized humidity. The values are divided to 100 (max)
	- windspeed: Normalized wind speed. The values are divided to 67 (max)
	- casual: count of casual users
	- registered: count of registered users
	- cnt: count of total rental bikes including both casual and registered

Task 2. (7 points)
Re-create the subplots below using:
- Variables: temp, cnt, season;
- Figsize = (10, 8);
- Title customization(hint: use fig.suptitle()): size is 14, weight is bold, position is (0.5,1.01); 
- x and y label customization: size is 12, weight is bold. Hint: use position=(0,-0.1) and position=(-0.1,0) to place the label into correct places;

```python
# do not run this line. Otherwise, the plot disappears
```

```python
df1 = day[day['season']==1]
df2 = day[day['season']==2]
df3 = day[day['season']==3]
df4 = day[day['season']==4]

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=[10, 8])
fig.suptitle('Bike Rentals at Different Temperatures By Season', x=0.5, y=1.01, fontsize=14, fontweight='bold')
fig.supxlabel('Normalized temperature', x=0.52, y=0.067, fontsize=12, fontweight='bold')
fig.supylabel('Count of Total Rental Bikes', x=0.055, y=0.5, fontsize=12, fontweight='bold')

ax[0, 0].scatter(x='temp', y='cnt', c='green', data=df1)
ax[0, 1].scatter(x='temp', y='cnt', c='orange', data=df2)
ax[1, 0].scatter(x='temp', y='cnt', c='brown', data=df3)
ax[1, 1].scatter(x='temp', y='cnt', c='blue', data=df4)

ax[0, 0].set_title('Spring', c='green', fontweight='bold')
ax[0, 1].set_title('Summer', c='orange', fontweight='bold')
ax[1, 0].set_title('Fall or Autumn', c='brown', fontweight='bold')
ax[1, 1].set_title('Winter', c='blue', fontweight='bold')

ax[0, 0].grid(True)
ax[0, 1].grid(True)
ax[1, 0].grid(True)
ax[1, 1].grid(True)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 7))
ax = sns.scatterplot(x='temp', y='cnt', data=day, hue='season', style='season',
                      palette=['green', 'orange', 'brown', 'blue'], legend='full')
ax.set_title('Bike Rentals at Different Temperatures \nBy Season', x=0.55, y=1.0, fontsize=16, 
                     fontweight='bold', ha='center')
ax.set_xlabel('Normalized Temperature', fontsize=14, fontweight='bold')
ax.set_ylabel('Count of Total Bike Rentals', fontsize=14, fontweight='bold')
plt.legend(loc='upper left', title='season', fontsize=11)
plt.show()
```
