# Customization

```python
fig, ax = plt.subplots()
ax.plot(df_usa['Year'], df_usa['gdp'])

#add labels to axis
ax.set_xlabel('Year')
ax.set_ylabel('GDP per capita')
#add title
ax.set_title('GDP per capita in the US')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(df_usa['Year'], df_usa['gdp'])

#change text size with fontdict
ax.set_xlabel('Year', fontdict = {'size':12})
ax.set_ylabel('GDP per capita', fontdict = {'size':12})
ax.set_title('GDP per capita in the US')
plt.show()
```

```python
fig, ax = plt.subplots()
#add marker 'circle'
#ax.plot(df_usa['Year'], df_usa['gdp'], marker = 'o')
#add marker 'thin_diamond'
ax.plot(df_usa['Year'], df_usa['gdp'], marker = 'd')

ax.set_xlabel('Year', fontdict = {'size':12})
ax.set_ylabel('GDP per capita', fontdict = {'size':12})
ax.set_title('GDP per capita in the US')
plt.show()
```

more info on markers here: https://matplotlib.org/stable/api/markers_api.html

```python
fig, ax = plt.subplots()
#add line style and color
ax.plot(df_usa['Year'], df_usa['gdp'], linestyle='--', color = 'g')

ax.set_xlabel('Year', fontdict = {'size':12})
ax.set_ylabel('GDP per capita', fontdict = {'size':12})
ax.set_title('GDP per capita in the US')
plt.show()
```

more_info on linestyles here: https://matplotlib.org/stable/api/_as_gen/matplotlib.lines.Line2D.html#matplotlib.lines.Line2D.set_linestyle

```python
fig, ax = plt.subplots()
ax.plot(df_usa['Year'], df_usa['gdp'], linestyle='--', color = 'g')
ax.set_xlabel('Year', fontdict = {'size':12})
ax.set_ylabel('GDP per capita', fontdict = {'size':12})
ax.set_title('GDP per capita in the US')
#showing grid
plt.grid(True)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(df_usa['Year'], df_usa['gdp'], linestyle='--', color = 'g')
#setting the xticks
ax.set_xticks(list(range(df_usa['Year'].min(), df_usa['Year'].max() + 1, 6)))
#setting the color and size
ax.set_xticklabels(list(range(df_usa['Year'].min(), df_usa['Year'].max() + 1, 6)), fontdict = {'size':12, 'color' : 'g'})

ax.set_xlabel('Year', fontdict = {'size':12})
ax.set_ylabel('GDP per capita', fontdict = {'size':12})
ax.set_title('GDP per capita in the US')
plt.grid(True)
plt.show()
```

```python
fig, ax = plt.subplots()
#add a legend to describe the variable that has been plotted
ax.plot(df_usa['Year'], df_usa['gdp'], linestyle='--', color = 'g', label = 'USA')
ax.set_xticks(list(range(df_usa['Year'].min(), df_usa['Year'].max() + 1, 6)))
ax.set_xticklabels(list(range(df_usa['Year'].min(), df_usa['Year'].max() + 1, 6)), fontdict = {'size':12, 'color' : 'g'})
ax.set_xlabel('Year', fontdict = {'size':12})
ax.set_ylabel('GDP per capita', fontdict = {'size':12})
ax.set_title('GDP per capita in the US')
ax.legend(loc = 'best')
plt.grid(True)
plt.show()
```

for more info on different legend: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html#

# Multiple Plots

```python
df_china = df[df['Country'] == 'China'].reset_index(drop = True)
df_china.shape
```

```python
fig, ax = plt.subplots()
ax.plot(df_usa['Year'], df_usa['gdp'], linestyle='-', color = 'b', label = 'USA')
ax.plot(df_china['Year'], df_china['gdp'], linestyle='-', color = 'r', label = 'China')
ax.set_xlabel('Year')
ax.set_ylabel('GDP per capita')
ax.legend(title = 'Country')
plt.show()
```

```python
#create multiple plots
fig, ax = plt.subplots(nrows = 2, ncols = 1, figsize = [12,10])
ax[0].plot(df_usa['Year'], df_usa['gdp'], linestyle='-', color = 'b', label = 'USA')
ax[1].plot(df_china['Year'], df_china['gdp'], linestyle='-', color = 'r', label = 'China')
ax[1].set_xlabel('Year')
ax[0].set_ylabel('GDP - USA')
ax[1].set_ylabel('GDP - China')
ax[0].set_title('Evolution of the GDP per capita: USA vs China')
ax[0].grid(True)
ax[1].grid(True)
plt.show()
```

```python
#create two series in the same plot
fig, ax = plt.subplots(figsize = [12,10])
ax.plot(df_usa['Year'], df_usa['gdp'], linestyle='-', color = 'b', label = 'USA')
ax.set_xlabel('Year')
ax.set_ylabel('GDP - USA', color = 'b')
ax.set_title('Evolution of the GDP per capita')
ax.tick_params('y', colors = 'b')

ax2 = ax.twinx()
ax2.plot(df_china['Year'], df_china['gdp'], linestyle='-', color = 'r', label = 'China')
ax2.set_ylabel('GDP - China', color = 'r')
ax2.tick_params('y', colors = 'r')
ax.grid(True)

lines, handles = ax.get_legend_handles_labels()
lines2, handles2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, handles + handles2, loc = 'best')

plt.show()
```

# Different Plot Types in Matplotlib

```python
import matplotlib.pyplot as plt
import pandas as pd
```

## Bar Charts

```python
df = pd.read_csv('fifa20_players_stats.csv', index_col = 0, usecols = [x for x in range(74)])
df.head()
```

```python
df.shape
```

```python
#the distribution of the nationalities among the players
#Which country has the higher number of football players?
count_countries = df[['Country', 'Name']].groupby('Country').count().rename(columns = {'Name':'Number of Players'})
count_countries_sorted = count_countries.sort_values('Number of Players', ascending = False)
count_countries_sorted.head(10)
```

```python
top10 = count_countries_sorted.head(10)
```

```python
fig, ax = plt.subplots(figsize = (10, 8))
ax.bar(top10.index, top10['Number of Players'], color = 'green')
ax.set_ylabel('Number of Players for each country')
#rotate the label's text
ax.set_xticklabels(top10.index, rotation = 90)
plt.show()
```

## Histograms

```python
#shows the distribution of values within a variable
```

```python
df.head()
```

```python
def normalize_data(df):
    normalized_df = (df['Overall'] - df['Overall'].mean())/df['Overall'].std()
    normalized_df = normalized_df[~normalized_df.isna()].to_frame()
    return normalized_df
```

```python
normalized_df = normalize_data(df)
```

```python
normalized_df.head()
```

```python
fig, ax = plt.subplots(figsize = (10, 8))
ax.hist(normalized_df['Overall'], label = 'Overall Skill')
ax.set_xlabel('Normalized Overall Skill')
ax.set_ylabel('Number of Observations')
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (10, 8))
#specify number of bars (default = 10)
ax.hist(normalized_df['Overall'], label = 'Overall Skill', bins = 70)
ax.set_xlabel('Normalized Overall Skill')
ax.set_ylabel('Number of Observations')
plt.show()
```

```python
england = df[df['Country'] == 'England'].reset_index(drop = True)
italy = df[df['Country'] == 'Italy'].reset_index(drop = True)
spain = df[df['Country'] == 'Spain'].reset_index(drop = True)
```

```python
eng_norm = normalize_data(england)
it_norm = normalize_data(italy)
sp_norm = normalize_data(spain)
```

```python
fig, ax = plt.subplots(nrows = 3, ncols = 1, figsize = (10, 8))
ax[0].hist(eng_norm['Overall'], label = 'Overall Skill', bins = 70)
ax[1].hist(it_norm['Overall'], label = 'Italy', bins = 70)
ax[2].hist(sp_norm['Overall'], label = 'Spain', bins = 70)

ax[2].set_xlabel('Normalized Overall Skill')
ax[1].set_ylabel('Number of Observations')
plt.show()
```

```python
#create three dataframes in the same plot, customize the overlapping
fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize = (10, 8))
ax.hist(eng_norm['Overall'], label = 'England', bins = 70, histtype = 'step')
ax.hist(it_norm['Overall'], label = 'Italy', bins = 70, histtype = 'step')
ax.hist(sp_norm['Overall'], label = 'Spain', bins = 70, histtype = 'step')

ax.set_xlabel('Normalized Overall Skill')
ax.set_ylabel('Number of Observations')
ax.legend()
plt.show()
```
