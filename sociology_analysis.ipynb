```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("sociology_survey.csv")
```

```python
df.head()
```

```python
df.drop(["Timestamp"], axis=1)
```

```python
df.rename({"Жасыңыз нешеде? / Сколько Вам лет?": "age"}, axis=1, inplace=True)
```

```python
df.rename({"Қай аймақта тұрасыз? / В каком регионе Вы проживаете?": "region"}, axis=1, inplace=True)
```

```python
df.rename({"Әлеуметтік желілерді қолданасыз ба? / Пользуетесь ли Вы социальными сетями? (Whatsapp, Instagram, Facebook, Telegram, Twitter, т.б./ и т.д)": "usage_of_social_networks"}, axis=1, inplace=True)
```

```python
df.rename({"Әлеуметтік желілерде күніне қанша сағат өткізесіз? / Сколько часов Вы проводите в социальных сетях?": "spent_time"}, axis=1, inplace=True)
```

```python
df.rename({"Қандай әлеуметтік желіде көп уақытыңызды өткізесіз? / В какой социальной сети Вы проводите больше всего времени?": "social_network_spent_most_time"}, axis=1, inplace=True)
```

```python
df.rename({"Пандемия басталғаннан кейін, Сізде әлеуметтік желілерде өткізу уақыты көбейді ме? / Увеличилось ли Ваше время проведения в социальных сетях с начала пандемии?": "Has your time on social media increased since the start of the pandemic?"}, axis=1, inplace=True)
```

```python
df.rename({"Әлеуметтік желілерді қолдану Сіздің оқу үлгеріміне әсер ете ме? / Сказывается ли использование социальных сетей на Вашей успеваемости в учебе?": "Does your use of social media affect your academic performance?"}, axis=1, inplace=True)
```

```python
df.rename({"Әлеуметтік желілерді көп уақытта қолдану Сіздің менталды денсаулығыңызға әсер ете ме? / Сказывается ли количество проведенного времени в социальных сетях на Вашем ментальном здоровье?": "Does the amount of time spent on social media affect your mental health?"}, axis=1, inplace=True)
```

```python
df.rename({"Әлеуметтік желілерде өткізетін уақытыңызды азайтқыңыз келе ме? / Хотели бы Вы уменьшить время, которое вы проводите в социальных сетях?": "How would you like to reduce the time you spend on social media?"}, axis=1, inplace=True)
```

```python
df.rename({"Сізде әлеуметтік желілерге құмарлық бар деп ойлайсыз ба? / Считаете ли Вы себя зависимым от социальных сетей?": "Do you consider yourself addicted to social networks?"}, axis=1, inplace=True)
```

```python
df.replace({"Батыс Қазақстан / Западный Казахстан": "West Kazakhstan",
           "Солтүстік Қазақстан / Северный Казахстан": "North Kazakhstan", 
           "Орталық Қазақстан / Центральный Казахстан": "Central Kazakhstan",
           "Шығыс Қазақстан / Восточный Казахстан": "East Kazakhstan",
           "Оңтүстік Қазақстан / Южный Казахстан": "South Kazakhstan",
           "Алматы": "Almaty",
           "Шымкент": "Shymkent"}, inplace=True)
```

```python
df.head()
```

```python
df.drop(["usage_of_social_networks"], axis=1, inplace=True)
```

```python
df.replace({"2-4 сағат / 2-4 часа": "2-4 hours", 
           "5-6 сағат / 5-6 часов": "5-6 hours",
           "1-2 сағат / 1-2 часа": "1-2 hours",
           "7-8 сағат / 7-8 часов": "7-8 hours",
           "8 сағаттан артық / больше 8 часов": "more than 8 hours"}, inplace=True)
```

```python
df.head()
```

```python
df.replace({"Иә / Да": "Yes", "Жоқ / Нет": "No"}, inplace=True)
```

```python
df.replace({"басқа / другое": "other"}, inplace=True)
```

```python
df.head()
```

```python
df.groupby(["age", "social_network_spent_most_time"])["social_network_spent_most_time"].count()
```

```python
df1 = df.groupby(["age", "social_network_spent_most_time"])["social_network_spent_most_time"].count().to_frame(name="count").reset_index()
```

```python
fig, ax = plt.subplots(ncols=2, figsize=(13, 6))
ax[0].pie(df1[df1["age"]=="before 18"]["count"], labels=df1[df1["age"]=="before 18"]["social_network_spent_most_time"], autopct='%1.1f%%', startangle=90, textprops={"fontsize": 14})
ax[0].axis('equal')
ax[0].set_title("before 18", fontweight="bold", fontsize=14)
ax[1].pie(df1[df1["age"]=="18-24"]["count"], labels=df1[df1["age"]=="18-24"]["social_network_spent_most_time"], autopct='%1.1f%%', startangle=90, textprops={"fontsize": 14})
ax[1].axis('equal')
ax[1].set_title("18-24", fontweight="bold", fontsize=14)
plt.subplots_adjust(wspace=0.7)
plt.show()
```

```python
df2 = df.groupby(["age", "spent_time"])["spent_time"].count().to_frame(name="count").reset_index()
```
